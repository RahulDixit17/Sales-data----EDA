```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
# To check the null value in Amount column with its respective columns, if possibile, we  fill the null values with exact value. 
df[['User_ID',
 'Cust_name',
 'Product_ID',
 'Gender',
 'Age Group',
 'Age',
 'Marital_Status',
 'State',
 'Zone',
 'Occupation',
 'Product_Category',
 'Orders',
 'Amount',
 'Status',
 'unnamed1']][df.Amount.isnull()]
```

```python
list(df)
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.shape
```

```python
# TO drop the null values
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# To check the datatype of any column
df['Amount'].dtypes
```

```python
# To change the datatype of any column
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
# To check the columns
df.columns
```

```python
# To rename any column
df.rename(columns={'Cust_name':'Customer_name'},inplace=True)
```

## Exploratory Data Analysis (EDA)

```python
df.columns
```

### Below count plot showing the Female purchase count is more than male counterparts.

```python
x = sns.countplot(x='Gender',data=df)

for i in x.containers:
    x.bar_label(i,color='red')
```

```python
df
```

```python
df.describe()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount','Orders'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen 
```

### We can see most of the buyers are females; females total purchasing amount are greater than men.

```python
# Bar plot:

ax = sns.barplot(x='Gender',y='Amount',data=sales_gen)
for i in ax.containers:
    ax.bar_label(i,color='green')
```

### Different age group order count. Max order given by people of age-groups 26-35 .

```python
x =sns.countplot(x='Age Group',data=df,hue ='Gender')
plt.ylabel("Orders Count",color='Black')
plt.xlabel("Orders Count",color='Black')

for i in x.containers:`
    x.bar_label(i,color='red')
```

# Total amount vs age-group

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)

sns.barplot(x= 'Age Group',y='Amount',data=sales_age)
```

# State 


```python
df.columns
```

```python
# Total no of orders from top 10 states

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

```

## No of orders from top 10 states

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

plt.figure(figsize=(17,5))

sales_state = sns.barplot(data=sales_state,x='State',y='Orders')
for i in sales_state.containers:
       sales_state.bar_label(i,color='red')
        
        
plt.xlabel("States",color='Green')
plt.ylabel("No of Orders",color='Green')
```

```python
#Total amount vs state from top 10 states

df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

### No of amounts collected from top 10 states:

```python
sales_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

plt.figure(figsize=(17,5))

sales_amount = sns.barplot(x='State',y='Amount',data=sales_amount)
```

## Marital Status

### Mostly married people did more shopping.

```python
plt.figure(figsize=(7,5))
t = sns.countplot(x='Marital_Status',data=df)

for i in t.containers:
     t.bar_label(i)
   
```

### We can see that most of the buyers are married(0) women and they have high purchasing power.

```python
sales_amount = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data = sales_amount, x ='Marital_Status',y='Amount' ,hue='Gender')
```

## Occupation -

### Most of the people who gave order, work in IT-Sector

```python
plt.figure(figsize=(19,5))
x = sns.countplot(x='Occupation', data=df)

for i in x.containers:
     x.bar_label(i)

```

## Occupation vs Amount barplot

### From below bar chart we can most of the buyers are working in IT, Healthcare and Aviation sector.


```python
sales_amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

x = sns.barplot(data = sales_amount, x ='Occupation',y='Amount')
```
